```python
def coleta_dados_apartamentos(bairro, pagina):

    from requests_html import HTMLSession
    session = HTMLSession()

    # link da imobiliaria
    LINK =  f"https://www.vivareal.com.br/aluguel/parana/curitiba/bairros/{bairro}/apartamento_residencial/#onde=Brasil,Paraná,Curitiba,Bairros,{bairro.capitalize()},,,,BR>Parana>NULL>Curitiba>Bairros>{bairro.capitalize()},,,&pagina={pagina}"

    # requisicao
    # response = requests.get(LINK)
    response = session.get(LINK)

    # declarando beautifulsoup
    site = bs4.BeautifulSoup(response.text, "html.parser")

    print(LINK)

    # selecionando classes dos imóveis
    imoveis = site.find_all('article', {'class': 'property-card__container js-property-card'})
    print(len(imoveis))

    # armazena dados dos apartamentos em um dataframe
    df_imoveis = pd.DataFrame()

    for imovel in imoveis:
        
        # criando dataframe auxiliar para armazenar dados do imóvel
        df_imoveis_aux = pd.DataFrame()
        df_imoveis_aux.loc[0, 'pagina'] = pagina
        df_imoveis_aux.loc[0, 'link'] = 'https://www.vivareal.com.br' + imovel.find_all('a')[0]['href']
        df_imoveis_aux.loc[0, 'titulo'] = imovel.find_all('span', {'class': 'property-card__title js-cardLink js-card-title'})[0].getText()
        df_imoveis_aux.loc[0, 'endereco'] = imovel.find_all('span', {'class': 'property-card__address'})[0].getText()
        df_imoveis_aux.loc[0, 'tipo_locacao'] = 'apartamento'
        df_imoveis_aux.loc[0, 'preco_aluguel'] = imovel.find_all('div', {'class': 'property-card__price js-property-card-prices js-property-card__price-small'})[0].getText()
        try:
            df_imoveis_aux.loc[0, 'preco_condominio'] = imovel.find_all('div', {'class': 'property-card__price-details--condo'})[0].getText()
        except:
            df_imoveis_aux.loc[0, 'preco_condominio'] = 'Condomínio R$ 0,00'
        df_imoveis_aux.loc[0, 'preco_iptu'] = np.nan
        df_imoveis_aux.loc[0, 'quartos'] = imovel.find_all('li', {'class': 'property-card__detail-item property-card__detail-room js-property-detail-rooms'})[0].getText()
        df_imoveis_aux.loc[0, 'vagas_garagem'] = imovel.find_all('li', {'class': 'property-card__detail-item property-card__detail-garage js-property-detail-garages'})[0].getText()
        df_imoveis_aux.loc[0, 'area'] = imovel.find_all('span', {'class': 'property-card__detail-value js-property-card-value property-card__detail-area js-property-card-detail-area'})[0].getText()
        df_imoveis_aux.loc[0, 'suites'] = imovel.find_all('li', {'class': 'property-card__detail-item property-card__detail-bathroom js-property-detail-bathroom'})[0].getText()

        # armazenando no dataframe principal
        df_imoveis = pd.concat([df_imoveis, df_imoveis_aux], axis = 0)

    # resetando index
    print(df_imoveis.shape)
    df_imoveis = df_imoveis.reset_index(drop=True)

    return df_imoveis

def descobre_quantas_paginas_tem(bairro):
    # link da imobiliaria
    pagina = 1
    LINK =  f"https://www.vivareal.com.br/aluguel/parana/curitiba/bairros/{bairro}/apartamento_residencial/#onde=Brasil,Paraná,Curitiba,Bairros,{bairro.capitalize()},,,,BR>Parana>NULL>Curitiba>Bairros>{bairro.capitalize()},,,&pagina={pagina}"
            #   https://www.vivareal.com.br/aluguel/parana/curitiba/bairros/{bairro}/apartamento_residencial/#onde=Brasil,Paran%C3%A1,Curitiba,Bairros,{bairro.capitalize()},,,,BR%3EParana%3ENULL%3ECuritiba%3EBarrios%3E{bairro.capitalize()},,,&pagina=5
    # requisicao
    response = requests.get(LINK)

    # declarando beautifulsoup
    site = bs4.BeautifulSoup(response.text, "html.parser")

    total_imoveis_por_pagina = len(site.find_all('article', {'class': 'property-card__container js-property-card'}))
    total_imovel_por_bairro = int(site.find_all('strong', {'class', 'results-summary__count js-total-records'})[0].getText().strip())

    return math.ceil(total_imovel_por_bairro/total_imoveis_por_pagina)


df_i = pd.DataFrame()
for page in range(1,descobre_quantas_paginas_tem('bigorrilho') + 1):
    df_aux = coleta_dados_apartamentos('bigorrilho', page)
    df_i = pd.concat([df_i, df_aux], axis = 0)

# resetando index
df_i = df_i.reset_index(drop=True)
df_i
```

```python
def descobre_quantas_paginas_tem():
    # link da imobiliaria
    LINK =  f"https://www.vivareal.com.br/aluguel/parana/curitiba/bairros/bigorrilho/apartamento_residencial/#onde=,Paran%C3%A1,Curitiba,Bairros,Bigorrilho,,,neighborhood,BR%3EParana%3ENULL%3ECuritiba%3EBarrios%3EBigorrilho,,,&pagina=1"

    # requisicao
    response = requests.get(LINK)

    # declarando beautifulsoup
    site = bs4.BeautifulSoup(response.text, "html.parser")

    total_imoveis_por_pagina = len(site.find_all('article', {'class': 'property-card__container js-property-card'}))
    total_imovel_por_bairro = int(site.find_all('strong', {'class', 'results-summary__count js-total-records'})[0].getText().strip())

    return math.ceil(total_imovel_por_bairro/total_imoveis_por_pagina)

descobre_quantas_paginas_tem()
```
